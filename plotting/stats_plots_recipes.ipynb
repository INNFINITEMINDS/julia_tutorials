# Statistical Plots

We can use pre-existing "recipes" for statistical concepts and datatypes commonly used in StatsBase.
Recipes is a "macro" (`@recipe`) which in combination with Plots.jl allow for reusable plots.

Some useful recipes live in the packeges [StatPlots.jl](https://github.com/JuliaPlots/StatPlots.jl) and [PlotRecipes.jl](https://github.com/JuliaPlots/PlotRecipes.jl)

## Imports

```julia
using StatPlots, Distributions, RDatasets
pyplot()
```

## Correlation of random variables

* corrplot()
* cornerplot()
* marginalhist()

### corrplot() and cornerplot()

```julia
M = randn(1000,4)
M[:,3] -= 0.7M[:,1] + 2
M[:,4] += M[:,1].^2
corrplot(M, label = ["x$i" for i=1:4])
```

```julia
cornerplot(M)
```

## Marginal Histogram

```julia
n = 1000
x = rand(Gamma(2), n)
y = -0.5x + randn(n)

marginalhist(x, y, fc=:plasma, bins=40)
```

## Marginal Histogram from DataFrames

```julia
iris = dataset("datasets","iris")
marginalhist(iris, :PetalLength, :PetalWidth)
```
